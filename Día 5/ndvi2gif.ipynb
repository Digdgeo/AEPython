```python
myclass = NdviSeasonality(roi=roi, sat='S2', key='max', periods=24,start_year=2018, end_year=2022, index='ndvi')
s2 = myclass.get_year_composite()

vizParams = {'bands': ['p2', 'p10', 'p15'], 'min': 0.1, 'max':0.8}
Map.addLayer(s2, vizParams, 'Median_EVI_AN')
```
